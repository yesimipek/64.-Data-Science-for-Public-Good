```
students = {"Jane":{"grade":90,"student_number":4012}, 
          "Pat":{"grade":81,"student_number":4052},
          "Denis":{"grade":70,"student_number":4089},
          "Max":{"grade":72,"student_number":4322},
          "Lara":{"grade":74,"student_number":4042}}
total=0
upperlist=[]
lowerlist=[]
samelist=[]
for person in students:
  total=total+students[person]['grade']
for person in students:
  if students[person]['grade']>(total/len(students)):
    upperlist.append(person)
  elif students[person]['grade']<(total/len(students)):
    lowerlist.append(person) 
  else:
    samelist.append(person)
print(total/len(students))
print("Grades of",' and '.join(upperlist),"are above the average.")
print("Grades of",' and '.join(lowerlist),"are below the average.")
print("Grades of",' and '.join(samelist),"are same with the average.")
```
